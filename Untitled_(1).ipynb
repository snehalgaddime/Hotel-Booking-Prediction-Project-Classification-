### Removing Outliers

```python
df.info()
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df1 = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df1
```

```python
df=remove_outliers_iqr(df,'number of weekend nights')
sns.boxplot(df['number of weekend nights'])
```

### Encoding

```python
df.describe(include='all').T
```

```python
df.info()
```

```python
# Booking_ID column is not going to give good information so better to drop it for good performance
df=df.drop('Booking_ID',axis=1)
```

```python
#type of meal feature
df['type of meal'].value_counts()
```

```python
# Using Label Encoding for Booking status as it has Ordinal values
df['type of meal'] = df['type of meal'].replace('Not Selected',0).replace('Meal Plan 1',1).replace('Meal Plan 2',2)
```

```python
df['type of meal'].value_counts()
```

```python
#room type feature
df['room type'].value_counts()
```

```python
# Using Label Encoding for room type as it has Ordinal values
df['room type'] = df['room type'].replace('Room_Type 1',1).replace('Room_Type 2',2).replace('Room_Type 3',3).replace('Room_Type 4',4).replace('Room_Type 5',5).replace('Room_Type 6',6).replace('Room_Type 7',7)
```

```python
df['room type'].value_counts()
```

```python
#market segment type
df['market segment type'].value_counts()
```

```python
#Using One-Hot Encoding for market segment type as it has Nominal values
A=pd.get_dummies(df['market segment type'], prefix='market segment type')
df=pd.concat([df, A], axis=1)
df=df.drop('market segment type',axis=1)
```

```python
#Target Feature Booking status
df['booking status'].value_counts()
```

```python
# using Label encoding for booking status as it has only 2 values
df['booking status'] = df['booking status'].replace('Canceled',0).replace('Not_Canceled',1)
```

```python
## Model Training
```

```python
x=df.drop('booking status',axis=1)
y=df["booking status"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#### Logistic Regression
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def evaluate_logistic_regression_model(model, X_test, y_test):
    # Make predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]  # Predict probabilities for positive class

    # Calculate evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Create dictionary of evaluation metrics
    evaluation_metrics = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'ROC AUC': roc_auc
    }

    return evaluation_metrics
```

```python
# Training
evaluate_logistic_regression_model(logistic_model,x_train,y_train)
```

```python
# Testing
evaluate_logistic_regression_model(logistic_model,x_test,y_test)
```

```python
# Standerization
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)

# Train a logistic regression model
logistic_model = LogisticRegression()
logistic_model.fit(X_train_scaled, y_train)

# Evaluate the training model
evaluation_metrics = evaluate_logistic_regression_model(logistic_model, X_train_scaled, y_train)
print("Evaluation metrics of Train data:")
for metric, value in evaluation_metrics.items():
    print(f"{metric}: {value}")

print(' ')

# Evaluate the testing model
evaluation_metrics = evaluate_logistic_regression_model(logistic_model, X_test_scaled, y_test)
print("Evaluation metrics of Test data:")
for metric, value in evaluation_metrics.items():
    print(f"{metric}: {value}")
```
